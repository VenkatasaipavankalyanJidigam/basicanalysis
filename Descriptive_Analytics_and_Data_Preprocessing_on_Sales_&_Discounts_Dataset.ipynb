###Standardization of Numerical Variables

###To standardise the numerical values we use z-scores. It brings all the numerical columns into same scale for easy interpretation by using the formula z = (x-mu)/sigma  where x is original value, mu is the mean  and sigma is standard deviation

```python

from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Identify numerical columns in the dataset

numeric_cols = ['Volume', 'Avg Price', 'Total Sales Value',
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Standardizing using sklearn
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# After standardization: Histograms
for col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df_scaled[col], kde=True, color='green')
    plt.title(f'Standardized : {col}')
    plt.tight_layout()
    plt.show()
```

## Conversion of Categorical Data into Dummy Variables

### We use One-Hot Encoding as most of the ML Algorithms do not support the non numerical data, They accept only numeric data directly so we encode the non numeric data to 0's and 1's based on the data

```python
# Selecting the categorical columns to encode
categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# Applying one-hot encoding technique
df_encoded = pd.get_dummies(df_scaled, columns=categorical_cols, drop_first=True)

# Showing a sample of the transformed dataset
df_encoded.head()
```
